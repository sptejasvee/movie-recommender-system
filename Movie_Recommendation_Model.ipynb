Data Cleaning

```python
import ast
```

```python
# 9. Function to extract the 'name' key from genres and keywords
def extract_names(text):
    names = []
    for i in ast.literal_eval(text):
        names.append(i['name'])
    return names

df['genres'] = df['genres'].apply(extract_names)
df['keywords'] = df['keywords'].apply(extract_names)
```

```python
# 10. Function to extract top 3 actors from cast
def extract_top_3_cast(text):
    names = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            names.append(i['name'])
            counter += 1
        else:
            break
    return names

df['cast'] = df['cast'].apply(extract_top_3_cast)
```

```python
# 11. Function to fetch director from crew
def fetch_director(text):
    names = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            names.append(i['name'])
            break
    return names

df['crew'] = df['crew'].apply(fetch_director)
```

```python
# 12. Remove spaces for all extracted lists and convert to string
def collapse_strings(lst):
    return " ".join([i.replace(" ", "") for i in lst])

df['genres'] = df['genres'].apply(collapse_strings)
df['keywords'] = df['keywords'].apply(collapse_strings)
df['cast'] = df['cast'].apply(collapse_strings)
df['crew'] = df['crew'].apply(collapse_strings)
```

```python
# 13. Combine all the processed text columns into a single string "tags" for each movie
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```

Text Preprocessing

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove special characters, numbers, and punctuation
    text = re.sub(r"[^a-z\s]", "", text)

    # Remove extra whitespace
    text = re.sub(r"\s+", " ", text).strip()

    # Tokenization
    words = text.split()

    # Remove stopwords and lemmatize
    words = [
        lemmatizer.lemmatize(word)
        for word in words
        if word not in stop_words
    ]
    return " ".join(words)

df['tags'] = df['tags'].apply(preprocess_text)
```

```python
# 14. Reset the index and map movie titles to their DataFrame index positions
df = df.reset_index(drop = True)
indices = pd.Series(df.index, index=df['title'].str.lower()).drop_duplicates()
```

Feature Engineering

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# 15. Convert the tags into a matrix of token counts
count_vectorizer = CountVectorizer(stop_words="english")
count_matrix = count_vectorizer.fit_transform(df['tags'])
```

Similarity Matrix Generation

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 16. Compute the cosine similarity matrix
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

Movie Recommendation Model

```python
def recommend(title, cosine_sim=cosine_sim, df=df, indices=indices, n=10):
    title_lower = title.lower()

    if title_lower not in indices:
        return f"Sorry, '{title}' was not found in the database."

    idx = indices[title_lower]

    # Get the pairwise similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores in descending order
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the most similar movies
    sim_scores = sim_scores[1:n+1]

    # Get the movie indices from the sorted and sliced list
    movie_indices = [i[0] for i in sim_scores]

    return df['title'].iloc[movie_indices]
```

Model Testing

```python
print(recommend('The Dark Knight Rises', n=5))
```
